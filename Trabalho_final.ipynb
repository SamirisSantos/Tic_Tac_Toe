```python
def entrada_jogador():
    jogada = input('Digite a sua jogada (X ou O): ').upper()
    while jogada not in ['X','O']:
        print('Jogada inválida. Tente novamente.')
        jogada = input('Digite a sua jogada (X ou O): ').upper()
    return jogada
```

```python
def escolha_jogador(tabuleiro):
    posicao = int(input('Digite a posição onde deseja jogar (1-9): '))
    while posicao < 1 or posicao > 9:
        print('Posição inválida. Tente novamente.')
        posicao = int(input('Digite a posição onde deseja jogar (1-9): '))
    return posicao
```

```python
# ve se a posicao esta livre
def verificar_espaco(tabuleiro, posicao):
    return tabuleiro[posicao] == " "


# poe o X ou o O no tabuleiro (so se a casa estiver livre)
def colocar_marcador(tabuleiro, marcador, posicao):
    if verificar_espaco(tabuleiro, posicao):
        tabuleiro[posicao] = marcador
        return True
    return False


# verifica se o tabuleiro ja esta cheio (para dar empate)
def verificar_tabuleiro_cheio(tabuleiro):
    for i in range(1, 10):
        if tabuleiro[i] == " ":
            return False 
    return True
```

```python
combinacoes = [
	[1, 2, 3],   
	[4, 5, 6],   
	[7, 8, 9],   
	[1, 4, 7],   
	[2, 5, 8],   
	[3, 6, 9],   
	[1, 5, 9],   
	[3, 5, 7],   
]
def verificar_vitoria(tabuleiro, marcador):
	for combinacao in combinacoes:
		if (tabuleiro[combinacao[0]] == marcador and
			tabuleiro[combinacao[1]] == marcador and
			tabuleiro[combinacao[2]] == marcador):
			return True, combinacao
	return False
```

# O loop principal

### O loop principal é o que junta todos os componentes e mantém o jogo a funcionar do início ao fim

-> O `loop_principal()` é o que junta todos os componentes e mantém o jogo afuncionar do início ao fim.Assenta em dois ciclos `while True` encaixados: o de fora para jogar várias partidas seguidas, o de dentro para os turnos de uma só partida.

-> No arranque de cada partida (ciclo de fora) preparo tudo: crio o `tabuleiro` com as posições vazias, decido quem começa com `escolher_primeiro()`, pergunto ao Jogador 1 o símbolo com `entrada_jogador()` e dou ao Jogador 2 o símbolo contrário.

-> Monto o dicionário `simbolos`, que liga cada jogador ao seu `X` ou `O`, para depois bastar escrever `simbolos[jogador_atual]` em vez de andar a verificar de quem é a vez.

-> Dentro do ciclo dos turnos, a cada volta limpo o ecrã, mostro o tabuleiro e indico de quem é a vez.Peço a jogada com `escolha_jogador()` e tento colocá-la com `colocar_marcador()`.

-> Se a posição for inválida ou já estiver ocupada, `colocar_marcador()` devolve `False` e o `continue` repete o mesmo turno **sem trocar de jogador** - quem jogou mal volta a jogar.

-> Se a jogada for válida, verifico o fim da partida: `verificar_vitoria()` deteta três em linha e `verificar_tabuleiro_cheio()` deteta empate. Qualquer um faz `break` e termina a partida. Só se o jogo continuar é que troco de jogador.

-> Quando a partida acaba, volto ao ciclo de fora e a função `repetir()` pergunta se querem jogar outra vez. Se sim, recomeça tudo com um tabuleiro limpo; se não, o `break` final termina o programa. Em resumo: o `continue` repete o turno, o `break` interno termina a partida e o `break` externo termina o jogo.

## Loop_Principal:


```python
def loop_principal():                                              
    while True:  # Loop externo: permite jogar varios jogos seguidos.
        
        tabuleiro = [" "] * 10   # Cria o tabuleiro com 10 posições vazias (índice 0 ignorado, usa-se 1 a 9)
                  
                                      
        jogador_atual = escolher_primeiro()  # Decide qual jogador começa e guarda em jogador_atual
        
        simbolo_j1 = entrada_jogador()   # Pergunta ao Jogador 1 qual símbolo quer (X ou O)
                                       
        if simbolo_j1 == "X": # Se o Jogador 1 escolheu X...
            simbolo_j2 = "O"  # então o Jogador 2 fica com O
        else:                 # Caso contrário (Jogador 1 escolheu O)...
            simbolo_j2 = "X"  # o Jogador 2 fica com X
            
        simbolos = {"Jogador 1": simbolo_j1, "Jogador 2": simbolo_j2}  # Dicionário que associa cada jogador ao seu símbolo
        
        
        while True:                        # Loop interno: controla os turnos do jogo
            clear_output(wait=True)        # Limpa o ecrã antes de mostrar o tabuleiro atualizado
            exibe_tabuleiro(tabuleiro)     # Mostra o estado atual do tabuleiro
            
            print(f"Jogador 1 fica com {simbolo_j1}")  # Informa o símbolo do Jogador 1
            print(f"Jogador 2 fica com {simbolo_j2}")  # Informa o símbolo do Jogador 2
            print(f"Vez do {jogador_atual} ({simbolos[jogador_atual]})")  # Indica de quem é a vez e o seu símbolo
            
            pos = escolha_jogador(tabuleiro)  # Pede ao jogador a posição (1 a 9) onde quer jogar
            
            mar = colocar_marcador(tabuleiro, simbolos[jogador_atual], pos)  # Tenta colocar o símbolo na posição; devolve True/False
            if not mar:
                continue # volta ao início do loop sem trocar de jogador
        
            if verificar_vitoria(tabuleiro, simbolos[jogador_atual]):  # Verifica se o jogador atual fez 3 em linha
                exibe_tabuleiro(tabuleiro)  # Mostra o tabuleiro final da vitória
                print(f"{jogador_atual} Ganhou!") # Anuncia o vencedor
                break # Termina a partida (sai do loop interno)
        
            if verificar_tabuleiro_cheio(tabuleiro): # Verifica se o tabuleiro está cheio sem vencedor
                exibe_tabuleiro(tabuleiro)# Mostra o tabuleiro final do empate
                print("Empate")# Anuncia empate
                break # Termina a partida (sai do loop interno)
        
            if jogador_atual == "Jogador 1": # Se a vez era do Jogador 1...
                jogador_atual = "Jogador 2"  # ...passa a vez ao Jogador 2
            else:  # Caso contrário...
                jogador_atual = "Jogador 1"  # ...passa a vez ao Jogador 1
                
        if not repetir():# Pergunta se querem jogar outra vez; se a resposta for não...
            break # ...sai do loop externo e termina o programa

loop_principal() # Inicia o jogo chamando a função principal 
```
