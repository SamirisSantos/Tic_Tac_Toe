## Tic Tac Toe


Tic Tac Toe (Jogo da velha) é um jogo para dois jogadores que se alternam a colocar um símbolo num tabuleiro 3×3. Um jogador usa X e o outro usa O. 
O objetivo é ser o primeiro a conseguir 3 símbolos seguidos na horizontal, vertical ou diagonal.

#### Requisitos:

* 2 jogadores devem conseguir jogar o jogo (ambos sentados no mesmo computador)
* O tabuleiro deve ser impresso sempre que um jogador faz uma jogada
* Devem conseguir aceitar a introdução da posição do jogador e colocar um símbolo no tabuleiro

```python
tabuleiro_teste = ['#','X','O','X','O','X','O','X','O','X']
```

```python
# Passo 4
# possibildades  de vitorias
combinacoes = [
	[0, 1, 2],
	[3, 4, 5],
	[6, 7, 8],
	[0, 3, 6],
	[1, 4, 7],
	[2, 5, 8],
	[0, 4, 8],
	[2, 4, 6],
]
def verificar_vitoria(tabuleiro, marcador):
	for combinacao in combinacoes:
		if (tabuleiro[combinacao[0]] == marcador and 
			tabuleiro[combinacao[1]] == marcador and 
			tabuleiro[combinacao[2]] == marcador):
			return True, combinacao
	return False
```

```python
#teste
print(verificar_vitoria(tabuleiro_teste, 'X'))
print(verificar_vitoria(tabuleiro_teste, 'O'))
```

```python
# passo 9
def repetir():
	while True:
		resposta = input("Queres jogar novamente? (s/n)").lower()
		if resposta == "s" or resposta == "sim":
			print('Bem-vindo ao Tic Tac Toe!')
			return True
		elif resposta == "n" or resposta == "nao":
			print("Obrigada pelo jogo")
			return False
		else:
			print("Resposta inválida! Escreve 's' ou 'n'")
```

```python
#teste
repetir()
```
